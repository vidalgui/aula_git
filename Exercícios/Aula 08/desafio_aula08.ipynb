## Exercícios:

### 2:
- Substituir a palavra "NES" por "Nintendinho" no nome dos consoles e deixar todos os nomes em maiúsculos

```python
df["Console Name"] = df["Console Name"].str.replace("NES", "Nintendinho")
df["Console Name"] = df["Console Name"].str.upper()
df
```

### 3:
- Filtre o nome dos consoles com data de release depois de 2010


```python
df[df["Released Year"] > 2010][["Console Name", "Released Year"]]
```

### 4:
- df.describe(), df.info() e substituição dos missing values por "missing"

```python
df.describe()
```

```python
df.info()
```

```python
df = df.fillna("missing")
df.head()
```

### 5:
- Filtrar os consoles que foram descontinuados menos de 2 anos após o lançamento

```python
df["Time Until Discontinuation"] = -1 # Caso o console não esteja descontinuado, atribui o valor dele como -1
mask = (df["Discontinuation Year"] != 0)
df.loc[mask, "Time Until Discontinuation"] = df["Discontinuation Year"] - df["Released Year"]
df.head()
```

```python
df[df["Time Until Discontinuation"] < 2][["Console Name", "Time Until Discontinuation"]]
```

1. Houve um Videogame descontinuado antes de seu lançamento
2. O único videogame descontinuado em menos de 2 anos foi o SNES Classic Edition
