## 1. Importando o Pandas

```python
import pandas as pd
```

## 2. Leitura dos arquivos

```python
df_consoles = pd.read_excel("best-selling game consoles (1).xlsx", "consoles")
df_company = pd.read_csv("company.txt")
```

## 3. Visualização dos Dataframes

```python
display(df_consoles.tail())
display(df_company.tail())
```

## 4. Seleção de colunas e tratamento de dados

### 4.1 - Groupby e Sorting

```python
# Seleciona as colunas de interesse e as agrupa por empresa, fazendo a contagem de ocorrência de consoles
consoles_por_companhia = df_consoles[["Company", "Console Name"]].groupby('Company', as_index=False).agg({"Console Name":"count"})

# Classifica pelo número de consoles, descendente
consoles_por_companhia_sorted = consoles_por_companhia.sort_values("Console Name", ascending=False)

consoles_por_companhia_sorted
```

### 4.2 - Merge com o DF de Market Values
Neste passo, optamos por fazer um *left join*

```python
companhias_marketvalue = consoles_por_companhia_sorted.merge(df_company, on="Company", how="left")
```

### 4.3 - Renomeando as colunas das células para satisfazer a saída esperada

```python
companhias_marketvalue = companhias_marketvalue.rename({"Console Name":"qtd"}, axis=1)
```
